```python
import pandas as pd
from datetime import datetime 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./uber_data.csv")
data.columns
```

```python

df = data[data['tpep_pickup_datetime'] != 0]
df['tout_time'] = df.apply(lambda row: (datetime.strptime(str(row.iloc[2]), '%Y-%m-%d %H:%M:%S') - \
                                        datetime.strptime(str(row.iloc[1]), '%Y-%m-%d %H:%M:%S') ).total_seconds()/60, axis = 1)
# print(df['tout_time'].head)
df.plot(
    kind='scatter',
    x= 'trip_distance',
    y='tout_time',
    color='red'
)
plt.title('tour_time')
plt.show()
# print(df.describe())
# types = df['tout_time'].nlargest(3)
# print(types)
# print(df['tout_time'].nsmallest(3))

# print(df.groupby('passenger_count'))
```
